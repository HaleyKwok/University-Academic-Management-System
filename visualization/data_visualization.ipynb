```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV files
departments = pd.read_csv('Department.csv')
courses = pd.read_csv('Course.csv')
instructors = pd.read_csv('Instructor.csv')
students = pd.read_csv('Student.csv')
sections = pd.read_csv('Section.csv')
classrooms = pd.read_csv('Classroom.csv')

# Example analysis #1: Budgets of each department
plt.figure(figsize=(12, 6))
plt.bar(departments['dept_name'], departments['budget'], color='skyblue')
plt.xlabel('Department')
plt.ylabel('Budget')
plt.title('Budget Allocation by Department')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Example analysis #3: Average salary of instructors by department
avg_salary = instructors.groupby('dept_name')['salary'].mean()
plt.figure(figsize=(12, 6))
avg_salary.plot(kind='bar', color='orange')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.title('Average Instructor Salary by Department')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
